```python
import pandas as pd
import numpy as np
from decimal import Decimal

#>>>>Change path>>>>>>
Data = pd.read_csv("/Users/qchen2/Desktop/QX600_DuplexSup_52C_01OCT2024_ClusterData.csv",skiprows=4,header=0,index_col=False)
#Add  delimiter=','
#>>>> Clean Data to the last column with value minus 5>>>>>
Data = Data[0:72]
#>>>>Change path>>>>>>
PlateDesign = pd.read_excel ("/Users/qchen2/Desktop/QX600_DuplexSup_Design.xlsx")

Data.insert(1,"Sample",'')
SampleID = ["SampleID"]
for i in range(0,len(Data)):
    WellID = [*Data.iloc[i,0]]
    wid_row = WellID[0]
    x = ord(wid_row) - 65
    y = int(WellID[1]+WellID[2])-1
    PlateDesign.iloc[x,y]
    SampleID.append(PlateDesign.iloc[x,y]) 
Data['Sample'] = SampleID[1:len(SampleID)]

Ch1_Mean_H2O = []
Ch1_STD_H2O = []

for row in Data.index:
    for col in Data.columns:
        if Data.loc[row, col] == 'H2O': 
        #and Data.loc[row,'Target 1'] == 0 and Data.loc[row,'Target 2'] == 0:
            Ch1_Mean_H2O.append(Data.loc[row, 'Ch1 Mean']) 
            Ch1_STD_H2O.append(Data.loc[row, 'Ch1 StdDev'])
Ch1_Mean_Value = [Decimal(d) for d in  Ch1_Mean_H2O]
Ch1_STD_Value = [Decimal(d) for d in  Ch1_STD_H2O]
CH1_MTH = (sum(Ch1_Mean_Value)/len(Ch1_Mean_Value)) + 6*(sum(Ch1_STD_Value)/len(Ch1_STD_Value))
print ('Manual Threshold for Channel 1 is:', CH1_MTH)

Ch2_Mean_H2O = []
Ch2_STD_H2O = []

for row in Data.index:
    for col in Data.columns:
        if Data.loc[row, col] == 'H2O': 
        #and Data.loc[row,'Target 1'] == 0 and Data.loc[row,'Target 2'] == 0:
            Ch2_Mean_H2O.append(Data.loc[row, 'Ch2 Mean']) 
            Ch2_STD_H2O.append(Data.loc[row, 'Ch2 StdDev'])
Ch2_Mean_Value = [Decimal(d) for d in  Ch2_Mean_H2O]
Ch2_STD_Value = [Decimal(d) for d in  Ch2_STD_H2O]
CH2_MTH = (sum(Ch2_Mean_Value)/len(Ch2_Mean_Value)) + 6*(sum(Ch2_STD_Value)/len(Ch2_STD_Value))
print ('Manual Threshold for Channel 2 is:', CH2_MTH)

Ch3_Mean_H2O = []
Ch3_STD_H2O = []

for row in Data.index:
    for col in Data.columns:
        if Data.loc[row, col] == 'H2O':
            #and Data.loc[row,'Target 3'] == 0 and Data.loc[row,'Target 2'] == 0:
            Ch3_Mean_H2O.append(Data.loc[row, 'Ch3 Mean']) 
            Ch3_STD_H2O.append(Data.loc[row, 'Ch3 StdDev'])
Ch3_Mean_Value = [Decimal(d) for d in  Ch3_Mean_H2O]
Ch3_STD_Value = [Decimal(d) for d in  Ch3_STD_H2O]
CH3_MTH = (sum(Ch3_Mean_Value)/len(Ch3_Mean_Value)) + 6*(sum(Ch3_STD_Value)/len(Ch3_STD_Value))
print ('Manual Threshold for Channel 3 is:', CH3_MTH)

Ch4_Mean_H2O = []
Ch4_STD_H2O = []

for row in Data.index:
    for col in Data.columns:
        if Data.loc[row, col] == 'H2O':
            #and Data.loc[row,'Target 4'] == 0 and Data.loc[row,'Target 2'] == 0:
            Ch4_Mean_H2O.append(Data.loc[row, 'Ch4 Mean']) 
            Ch4_STD_H2O.append(Data.loc[row, 'Ch4 StdDev'])
Ch4_Mean_Value = [Decimal(d) for d in  Ch4_Mean_H2O]
Ch4_STD_Value = [Decimal(d) for d in  Ch4_STD_H2O]
CH4_MTH = (sum(Ch4_Mean_Value)/len(Ch4_Mean_Value)) + 6*(sum(Ch4_STD_Value)/len(Ch4_STD_Value))
print ('Manual Threshold for Channel 4 is:', CH4_MTH)

Ch5_Mean_H2O = []
Ch5_STD_H2O = []

for row in Data.index:
    for col in Data.columns:
        if Data.loc[row, col] == 'H2O':
            #and Data.loc[row,'Target 5'] == 0 and Data.loc[row,'Target 2'] == 0:
            Ch5_Mean_H2O.append(Data.loc[row, 'Ch5 Mean']) 
            Ch5_STD_H2O.append(Data.loc[row, 'Ch5 StdDev'])
Ch5_Mean_Value = [Decimal(d) for d in  Ch5_Mean_H2O]
Ch5_STD_Value = [Decimal(d) for d in  Ch5_STD_H2O]
CH5_MTH = (sum(Ch5_Mean_Value)/len(Ch5_Mean_Value)) + 6*(sum(Ch5_STD_Value)/len(Ch5_STD_Value))
print ('Manual Threshold for Channel 5 is:', CH5_MTH)

Ch6_Mean_H2O = []
Ch6_STD_H2O = []

for row in Data.index:
    for col in Data.columns:
        if Data.loc[row, col] == 'H2O': 
        #and Data.loc[row,'Target 6'] == 0 and Data.loc[row,'Target 2'] == 0:
            Ch6_Mean_H2O.append(Data.loc[row, 'Ch6 Mean']) 
            Ch6_STD_H2O.append(Data.loc[row, 'Ch6 StdDev'])
Ch6_Mean_Value = [Decimal(d) for d in  Ch6_Mean_H2O]
Ch6_STD_Value = [Decimal(d) for d in  Ch6_STD_H2O]
CH6_MTH = (sum(Ch6_Mean_Value)/len(Ch6_Mean_Value)) + 6*(sum(Ch6_STD_Value)/len(Ch6_STD_Value))
print ('Manual Threshold for Channel 6 is:', CH6_MTH)

#>>>>>>> Activate code to puke file>>>>>>>>>>
#Data.to_excel('VacA_Plate2_ClusterData_WithID.xlsx', index=False)
```
