```python
### FUNCTION: Install and import libraries 
#%pip install openpyxl
#%pip install pandas
import pandas as pd
import numpy as np
from decimal import Decimal

### FUNCTION: Import files and clean data
################ ACTION NEEDED: Change the path to your file location
PlateDesign = pd.read_excel ("/Users/qchen2/Desktop/6X_Plate1_Design.xlsx")
Data = pd.read_csv("/Users/qchen2/Desktop/6X_Plate1_ClusterData.csv", skiprows=4,header=0,index_col=False)
# Forcing the 0 and 1 in these columns to be read as numbers instead of text
for i in range(1, 8):
    Data.iloc[:, i] = pd.to_numeric(Data.iloc[:, i], errors='coerce')
# End before the first irrelevant row in QXmanager files. 
# No need if your file is from Quantasoft.
Row_Starts_With_Well = Data[Data.iloc[:, 0].str.startswith('Well', na=False)]
Row_index = Row_Starts_With_Well.index[0]
Data = Data[0:Row_index]

### FUNCTION: Fill ID Names based on their location on platedeisgn
# In the first column of Data, insert a column called Sample
Data.insert(1,"Sample",'')
SampleID = ["SampleID"]
# For the location of well, ex: A01, break it into A and 01
# A as 1 on X axis and 01 as 1 on the Y axis. A01 = (1, 1)
# Find (X, Y) on the platedesign and eppend the value of that cell (the matching sampleID) to this well location. 
for i in range(0,len(Data)):
    WellID = [*Data.iloc[i,0]]
    wid_row = WellID[0]
    x = ord(wid_row) - 65
    y = int(WellID[1]+WellID[2])-1
    PlateDesign.iloc[x,y]
    SampleID.append(PlateDesign.iloc[x,y]) 
Data['Sample'] = SampleID[1:len(SampleID)]

### FUNCTION: Export file with SampleID filled. The file will be in the same location you host these codes. 
################ ACTION NEEDED: Change the name of the file you want to export
Data.to_excel('6X_Plate1_ClusterData_WithID.xlsx', index=False)

### FUNCTION: Calculate Manual Threshold for Channel 1.
# Create list to record the values of Mean and Standard deviation of H2O. 
Ch1_Mean_H2O = []
Ch1_STD_H2O = []
# Find all the values matching to samples named "H2O"
for row in Data.index:
    for col in Data.columns:
         if Data.loc[row, col] == 'H2O' 
         # Activate following code if you only want 0 in water. 
         # and Data.loc[row,'Target 1'] == 0 and Data.loc[row,'Target 2'] == 0 and Data.loc[row,'Target 3'] == 0 and Data.loc[row,'Target 4'] == 0 and Data.loc[row,'Target 5'] == 0 and Data.loc[row,'Target 6'] == 0:
            Ch1_Mean_H2O.append(Data.loc[row, 'Ch1 Mean']) 
            Ch1_STD_H2O.append(Data.loc[row, 'Ch1 StdDev'])
# Allows decimal for calculation
Ch1_Mean_Value = [Decimal(d) for d in  Ch1_Mean_H2O]
Ch1_STD_Value = [Decimal(d) for d in  Ch1_STD_H2O]
# Calculate the Manual Threshold
CH1_MTH = (sum(Ch1_Mean_Value)/len(Ch1_Mean_Value)) + 6*(sum(Ch1_STD_Value)/len(Ch1_STD_Value))
print ('Manual Threshold for Channel 1 is:', CH1_MTH)

### FUNCTION: Calculate Manual Threshold for Channel 2.
# Since the codes are the same as channel 1, I am not repeating the comment. 
# If uncleared, refer to channel 1 comments for clarification. 
Ch2_Mean_H2O = []
Ch2_STD_H2O = []
for row in Data.index:
    for col in Data.columns:
         if Data.loc[row, col] == 'H2O' 
         # and Data.loc[row,'Target 1'] == 0 and Data.loc[row,'Target 2'] == 0 and Data.loc[row,'Target 3'] == 0 and Data.loc[row,'Target 4'] == 0 and Data.loc[row,'Target 5'] == 0 and Data.loc[row,'Target 6'] == 0:
            Ch2_Mean_H2O.append(Data.loc[row, 'Ch2 Mean']) 
            Ch2_STD_H2O.append(Data.loc[row, 'Ch2 StdDev'])
Ch2_Mean_Value = [Decimal(d) for d in  Ch2_Mean_H2O]
Ch2_STD_Value = [Decimal(d) for d in  Ch2_STD_H2O]
CH2_MTH = (sum(Ch2_Mean_Value)/len(Ch2_Mean_Value)) + 6*(sum(Ch2_STD_Value)/len(Ch2_STD_Value))
print ('Manual Threshold for Channel 2 is:', CH2_MTH)

### FUNCTION: Calculate Manual Threshold for Channel 3.
# Since the codes are the same as channel 1, I am not repeating the comment. 
# If uncleared, refer to channel 1 comments for clarification. 
Ch3_Mean_H2O = []
Ch3_STD_H2O = []
for row in Data.index:
    for col in Data.columns:
         if Data.loc[row, col] == 'H2O' 
         # and Data.loc[row,'Target 1'] == 0 and Data.loc[row,'Target 2'] == 0 and Data.loc[row,'Target 3'] == 0 and Data.loc[row,'Target 4'] == 0 and Data.loc[row,'Target 5'] == 0 and Data.loc[row,'Target 6'] == 0:
            Ch3_Mean_H2O.append(Data.loc[row, 'Ch3 Mean']) 
            Ch3_STD_H2O.append(Data.loc[row, 'Ch3 StdDev'])
Ch3_Mean_Value = [Decimal(d) for d in  Ch3_Mean_H2O]
Ch3_STD_Value = [Decimal(d) for d in  Ch3_STD_H2O]
CH3_MTH = (sum(Ch3_Mean_Value)/len(Ch3_Mean_Value)) + 6*(sum(Ch3_STD_Value)/len(Ch3_STD_Value))
print ('Manual Threshold for Channel 3 is:', CH3_MTH)

### FUNCTION: Calculate Manual Threshold for Channel 4.
# Since the codes are the same as channel 1, I am not repeating the comment. 
# If uncleared, refer to channel 1 comments for clarification. 
Ch4_Mean_H2O = []
Ch4_STD_H2O = []
for row in Data.index:
    for col in Data.columns:
         if Data.loc[row, col] == 'H2O' 
         # and Data.loc[row,'Target 1'] == 0 and Data.loc[row,'Target 2'] == 0 and Data.loc[row,'Target 3'] == 0 and Data.loc[row,'Target 4'] == 0 and Data.loc[row,'Target 5'] == 0 and Data.loc[row,'Target 6'] == 0:
            Ch4_Mean_H2O.append(Data.loc[row, 'Ch4 Mean']) 
            Ch4_STD_H2O.append(Data.loc[row, 'Ch4 StdDev'])
Ch4_Mean_Value = [Decimal(d) for d in  Ch4_Mean_H2O]
Ch4_STD_Value = [Decimal(d) for d in  Ch4_STD_H2O]
CH4_MTH = (sum(Ch4_Mean_Value)/len(Ch4_Mean_Value)) + 6*(sum(Ch4_STD_Value)/len(Ch4_STD_Value))
print ('Manual Threshold for Channel 4 is:', CH4_MTH)

### FUNCTION: Calculate Manual Threshold for Channel 5.
# Since the codes are the same as channel 1, I am not repeating the comment. 
# If uncleared, refer to channel 1 comments for clarification. 
Ch5_Mean_H2O = []
Ch5_STD_H2O = []
for row in Data.index:
    for col in Data.columns:
         if Data.loc[row, col] == 'H2O' 
         # and Data.loc[row,'Target 1'] == 0 and Data.loc[row,'Target 2'] == 0 and Data.loc[row,'Target 3'] == 0 and Data.loc[row,'Target 4'] == 0 and Data.loc[row,'Target 5'] == 0 and Data.loc[row,'Target 6'] == 0:
            Ch5_Mean_H2O.append(Data.loc[row, 'Ch5 Mean']) 
            Ch5_STD_H2O.append(Data.loc[row, 'Ch5 StdDev'])
Ch5_Mean_Value = [Decimal(d) for d in  Ch5_Mean_H2O]
Ch5_STD_Value = [Decimal(d) for d in  Ch5_STD_H2O]
CH5_MTH = (sum(Ch5_Mean_Value)/len(Ch5_Mean_Value)) + 6*(sum(Ch5_STD_Value)/len(Ch5_STD_Value))
print ('Manual Threshold for Channel 5 is:', CH5_MTH)

### FUNCTION: Calculate Manual Threshold for Channel 6.
# Since the codes are the same as channel 1, I am not repeating the comment. 
# If uncleared, refer to channel 1 comments for clarification. 
Ch6_Mean_H2O = []
Ch6_STD_H2O = []
for row in Data.index:
    for col in Data.columns:
       if Data.loc[row, col] == 'H2O' 
       # and Data.loc[row,'Target 1'] == 0 and Data.loc[row,'Target 2'] == 0 and Data.loc[row,'Target 3'] == 0 and Data.loc[row,'Target 4'] == 0 and Data.loc[row,'Target 5'] == 0 and Data.loc[row,'Target 6'] == 0:
            Ch6_Mean_H2O.append(Data.loc[row, 'Ch6 Mean']) 
            Ch6_STD_H2O.append(Data.loc[row, 'Ch6 StdDev'])
Ch6_Mean_Value = [Decimal(d) for d in  Ch6_Mean_H2O]
Ch6_STD_Value = [Decimal(d) for d in  Ch6_STD_H2O]
CH6_MTH = (sum(Ch6_Mean_Value)/len(Ch6_Mean_Value)) + 6*(sum(Ch6_STD_Value)/len(Ch6_STD_Value))
print ('Manual Threshold for Channel 6 is:', CH6_MTH)
```
