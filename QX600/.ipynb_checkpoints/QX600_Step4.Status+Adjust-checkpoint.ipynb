```python
#####UNFINISHED#####STOPED AT STATUS because Nina says so. 
import pandas as pd
import numpy as np
MasterList = pd.read_excel ("/Users/qchen2/Desktop/MasterList_Filled.xlsx")

# COLUMN G6 for Status of HP 16S
# HP 16S	0,1	0= Negative, 1=Positive
for x in range(2, len(MasterList)):
    if MasterList.iloc[x,12] >= 5:
        MasterList.iloc[x,6] = 1
    if MasterList.iloc[x,12] < 5:
        MasterList.iloc[x,6] = 0

# COLUMN H7 for Status of CagA
#CagA	0,1	0= Negative, 1=Positive
for x in range(2, len(MasterList)):
    if MasterList.iloc[x,13] >= 5:
        MasterList.iloc[x,7] = 1
    if MasterList.iloc[x,13] < 5:
        MasterList.iloc[x,7] = 0

# COLUMN I8 for Status of EPIYA
#EPIYA	0,1,2,3	0=Negative, 1=EPIYA C, 2=EPIYA D, 3=EPIYA Mixed
for x in range(2, len(MasterList)):
    if MasterList.iloc[x,14] >= 5 and MasterList.iloc[x,15] >= 5:
        MasterList.iloc[x,8] = 3
    if MasterList.iloc[x,14] >= 5 and MasterList.iloc[x,15] < 5:
        MasterList.iloc[x,8] = 1
    if MasterList.iloc[x,14] < 5 and MasterList.iloc[x,15] >= 5:
        MasterList.iloc[x,8] = 2
    if MasterList.iloc[x,14] < 5 and MasterList.iloc[x,15] < 5:
        MasterList.iloc[x,8] = 0

# COLUMN K10 for Status of VacA 
# VacA	0,1,2,3	0=Negative, 1=VacA i1, 2=VacA i2, 3=VacA Mixed
for x in range(2, len(MasterList)):
    if MasterList.iloc[x,16] >= 5 and MasterList.iloc[x,17] >= 5:
        MasterList.iloc[x,10] = 3
    # IF VacA i1 Droplets >= 5 and VacA i2 Droplets <5: Status = 1
    if MasterList.iloc[x,16] >= 5 and MasterList.iloc[x,17] < 5:
        MasterList.iloc[x,10] = 1
    # IF VacA i1 Droplets < 5 and VacA i2 Droplets >=5: Status = 2
    if MasterList.iloc[x,16] < 5 and MasterList.iloc[x,17] >= 5:
        MasterList.iloc[x,10] = 2
    # IF VacA i1 Droplets <5 and VacA i2 Droplets <5: Status = 0
    if MasterList.iloc[x,16] < 5 and MasterList.iloc[x,17] < 5:
        MasterList.iloc[x,10] = 0

# COLUMN J for %EPIYA

# COLUMN V for Adjusted Hp 16S Droplets 
# COLUMN W for Adjusted cagA Droplets 
# COLUMN X for Adjusted EPIYA-C Droplets 
# COLUMN Y for Adjusted EPIYA-D Droplets 
# COLUMN Z for Adjusted VacAi1 Droplets 
# COLUMN AA for Adjusted VacAi2 Droplets 
# COLUMN AD for Adjusted 18S Droplets 

# COLUMN AN for Adjusted Hp 16S Concentration
# COLUMN AO for Adjusted cagA Concentration
# COLUMN AP for Adjusted EPIYA-C Concentration
# COLUMN AQ for Adjusted EPIYA-D Concentration 
# COLUMN AR for Adjusted VacAi1 Concentration
# COLUMN AS for Adjusted VacAi2 Concentration
# COLUMN AV for Adjusted 18S Concentration

# COLUMN AW for Copy # of Hp16S in 10 ul input
# COLUMN AX for Copy # of cagA in 10 ul input

MasterList.columns = ["" if "Unnamed" in str(col) else col for col in MasterList.columns]
MasterList.to_excel('MasterList_6X_06DEC2024.xlsx', index=False)
```
