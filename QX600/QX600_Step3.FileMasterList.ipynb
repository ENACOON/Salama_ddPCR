```python
### FUNCTION: Install and import libraries 
import pandas as pd
import numpy as np

### FUNCTION: Import files and clean data
################ ACTIONS NEEDED: Change the path to your file location>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
ANA = pd.read_excel ("/Users/qchen2/Desktop/6X_Plate1_MTH_WithID_Averaged.xlsx")
MasterList = pd.read_excel ("/Users/qchen2/Desktop/MasterList.xlsx")
# I don't want the empty rolls. So this line drops them. 
AVRS = ANA.dropna()
AVRS = AVRS[['Sample', 'Target', 'Average Conc (copies/uL)','Average Positives']]

### FUNCTION: Assign names of probes to the channels. 
# channel 1 - FAM = CagA
# channel 3 - Cy5 = 16S
# channel 4 - Cy5.5 = EPIYA_C
# channel 6 - ATTO590 = EPIYA_D
# channel 2 - HEX = VacA_i1
# channel 5 - ROX = VacA_i2
avrs_FAM = AVRS[AVRS['Target'] == 1]
avrs_Cy5 = AVRS[AVRS['Target'] == 3]
avrs_Cy5_5 = AVRS[AVRS['Target'] == 4]
avrs_ATTO590 = AVRS[AVRS['Target'] == 6]
avrs_HEX = AVRS[AVRS['Target'] == 2]
avrs_ROX = AVRS[AVRS['Target'] == 5]

# This is a safeguard to catch anything wrong. 
# If a match based on sampleID is not found comparing the imput data to masterlist, it will print the sampleID out. 
unmatched_values = []

#FILL channel 1 - FAM = CagA
for x1, y1 in avrs_FAM.iterrows():
    for x, y in MasterList.iterrows():
        found_match = False 
        # Compare the SampleID from Data File to the SampleID on MasterList
        if y1.iloc[0] == y.iloc[0]:
################ ACTIONS NEEDED: Change the well location based on your MasterList >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
            #When there is a match, record the value of averaged positive droplets to the corresponding well location (column N, row of the SampleID). 
            y.iloc[13] = y1.iloc[3]# COLUMN N for Positive Droplets of CagA
            #When there is a match, record the value of averaged concentration to the corresponding well location (column AF, row of the SampleID). 
            y.iloc[31] = y1.iloc[2] # COLUMN AF for Concentration of VacA i2 y2.iloc[35]
            found_match = True
            break
    # If there isn't a match, record the SampleID to the unmatched value list. 
    if not found_match:
        unmatched_values.append(y1.iloc[0])
        
#FILL channel 2 - HEX = VacA_i1
for x2, y2 in avrs_HEX.iterrows():
    for x, y in MasterList.iterrows():
        found_match = False 
        if y2.iloc[0] == y.iloc[0]:
################ ACTIONS NEEDED: Change the well location based on your MasterList >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
            y.iloc[16] = y2.iloc[3] # COLUMN Q for Positive Droplets of VacA_i1
            y.iloc[34] = y2.iloc[2]  # COLUMN AI for Concentration of VacA i1 
            found_match = True
            break
    if not found_match:
        unmatched_values.append(y2.iloc[0])
        
#FILL channel 3 - Cy5 = 16S
for x3, y3 in avrs_Cy5.iterrows():
    for x, y in MasterList.iterrows():
        found_match = False 
        if y3.iloc[0] == y.iloc[0]:
 ################ ACTIONS NEEDED: Change the well location based on your MasterList >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
            y.iloc[12] = y3.iloc[3] # COLUMN M for Positive Droplets of Hp16S
            y.iloc[30] = y3.iloc[2]  # COLUMN AE for Concentration of Hp16S
            found_match = True
            break
    if not found_match:
        unmatched_values.append(y3.iloc[0])

#FILL channel 4 - Cy5.5 = EPIYA_C
for x4, y4 in avrs_Cy5_5.iterrows():
    for x, y in MasterList.iterrows():
        found_match = False 
        if y4.iloc[0] == y.iloc[0]:
 ################ ACTIONS NEEDED: Change the well location based on your MasterList >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
            y.iloc[14] = y4.iloc[3] # COLUMN O for Positive Droplets of EPIYA-C
            y.iloc[32] = y4.iloc[2]  # COLUMN AG for Concentration of EPIYA-C
            found_match = True
            break
    if not found_match:
        unmatched_values.append(y4.iloc[0])

#FILL channel 5 - ROX = VacA_i2
for x5, y5 in avrs_ROX.iterrows():
    for x, y in MasterList.iterrows():
        found_match = False 
        if y5.iloc[0] == y.iloc[0]:
################ ACTIONS NEEDED: Change the well location based on your MasterList >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
            y.iloc[17] = y5.iloc[3] # COLUMN R for Positive Droplets of VacA_i2
            y.iloc[35] = y5.iloc[2]  # COLUMN AJ for Concentration of VacA_i2
            found_match = True
            break
    if not found_match:
        unmatched_values.append(y5.iloc[0])


#FILL channel 6 - ATTO590 = EPIYA_D
for x6, y6 in avrs_ATTO590.iterrows():
    for x, y in MasterList.iterrows():
        found_match = False 
        if y6.iloc[0] == y.iloc[0]:
 ################ ACTIONS NEEDED: Change the well location based on your MasterList >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
            y.iloc[15] = y6.iloc[3] # COLUMN P for Positive Droplets of EPIYA-D
            y.iloc[33] = y6.iloc[2]  # COLUMN AH for Concentration of EPIYA-D
            found_match = True
            break
    if not found_match:
        unmatched_values.append(y6.iloc[0])

# Check the print out. It should only contain your controls (H2O, etc.) and none of your samples. 
print ("These samples are not found in the MasterList:", unmatched_values)

################ ACTION NEEDED: Change the name of the file you want to export. >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
MasterList.columns = ["" if "Unnamed" in str(col) else col for col in MasterList.columns]
MasterList.to_excel('MasterList_Filled.xlsx', index=False)
```
